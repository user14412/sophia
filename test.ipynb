```python
import torch
import torchaudio
import ChatTTS
import os
import inspect

chat = ChatTTS.Chat()

model_path = "./model/ChatTTS/ChatTTS"

if not os.path.exists(model_path):
    print(f"错误：找不到模型路径 {model_path}")

chat.load(source='local', local_path=model_path, compile=False)

texts = ["你好，欢迎使用ChatTTS！"]

wavs = chat.infer(texts)

torchaudio.save("output.wav", torch.from_numpy(wavs[0]), 24000) # 路径，源张量，采样率
```
